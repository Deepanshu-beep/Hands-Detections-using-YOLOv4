### Created function to detect hands for Realtime/ Video Input

```python
def detectHands(net, frame, ln):
    results=[]
    (H,W)=frame.shape[:2]
    blob=cv2.dnn.blobFromImage(frame, 1/255.0, (416,416), swapRB=True, crop=False)
    net.setInput(blob)
    layerOutput=net.forward(ln)
    boxes=[]
    centers=[]
    confidences=[]
    for outputs in layerOutput:
        for detections in outputs:
            scores=detections[5:]
            classID=np.argmax(scores)
            confidence=scores[classID]
            box=detections[:4]*np.array([W,H,W,H])
            (centerX, centerY, width, height)=box.astype('int')
            topLeftX=int(centerX-(width/2))
            topLeftY=int(centerY-(height/2))
            boxes.append([topLeftX, topLeftY, int(width), int(height)])
            centers.append((centerX, centerY))
            confidences.append(float(confidence))
    non_max_suppression=cv2.dnn.NMSBoxes(boxes, confidences, min_Conf, NMS_thresh)
    if len(non_max_suppression)>0:
        for i in non_max_suppression.flatten():
            (x1, y1, x2, y2)=(boxes[i][0], boxes[i][1], boxes[i][0]+boxes[i][2], boxes[i][1]+boxes[i][3])
            temp=(confidences[i], (x1, y1, x2, y2), centers[i])
            results.append(temp)
    return results
```

```python
#Minimum confidence to filter weak detections

min_Conf=0.9

#Minimum confidence for Non-Max Suppression to remove extra detections of boxes

NMS_thresh=0.99
writer=None
```

```python
# vs=cv2.VideoCapture(0)
vs=cv2.VideoCapture(0)
annotations=[]
while True:
    check, frame=vs.read()
    if not check:
        break
    frame=imutils.resize(frame, width=700)
    results=detectHands(model, frame, ln)
    for (i, (conf, box, centers)) in enumerate(results):
        (centerX, centerY)=centers
#         (x1,x2, x3,x4)=box
#         cv2.rectangle(frame, (x1, x2), (x3, x4), (0,255,0),2)
        annotations.append(centers)
    if len(annotations)>1:
        for i in range(len(annotations)-1):
            cv2.line(frame, annotations[i], annotations[i+1],(0,255,0),2)
#     cv2.imshow('Frame', frame)
#     key=cv2.waitKey(10)
#     if key==27:
#         break
    if writer==None and output!='':
        fourcc=cv2.VideoWriter_fourcc(*'MJPG')
        writer=cv2.VideoWriter(output, fourcc, 10, (frame.shape[1],frame.shape[0]),True)
    if writer is not None:
        writer.write(frame)
writer.release()
# vs.release()
# cv2.destroyAllWindows()
```
